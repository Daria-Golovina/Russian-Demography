Временная миграция (или же вся миграция, включая временную) может послужить темой отдельного исследования - если же мы хотим попробовать оценить постоянную миграцию, то есть "переезд навсегда", имеет смысл рассмотреть графу `прибыли и зарегистрированы в новом месте жительства`. Начнём с Санкт-Петербурга - выделим данные, относящиеся к нему:

```python
data_spb = data_migration[data_migration['region'] == 'Санкт-Петербург'].copy()
data_spb_in = data_spb[data_spb['registration_type'] == 'прибыли и зарегистрированы в новом месте жительства']
display(data_spb_in)
```

Получилось - вырежем всё лишнее и построим график переехавших по годам:

### <center>Основные места постоянной миграции</center>

Посмотрим, куда люди в основном переезжают - выделим нужную графу и проверим, есть ли пропуски в нужных нам `indicator_value`:

```python
data_top_in = data_migration[(data_migration['registration_type'] == 'прибыли и зарегистрированы в новом месте жительства') & 
                            (data_migration['year'] >= 2011)].copy()
print(data_top_in.info())
```

Все `indicator_value` на месте - можно собирать значения в таблицу. Также вспомним про федеральные округа и общее значение `Российская Федерация` и уберём их:

```python
#сбор в таблицу
top_in = pd.pivot_table(data_top_in, index='region', values='indicator_value', aggfunc='sum').sort_values(
    by='indicator_value', ascending=False)
top_in['indicator_value'] = top_in['indicator_value'].astype('int64')

#убираем фед. округа и общую сумму
top_in = top_in[(top_in.index.str.contains('федеральный округ') == False) & 
                      (top_in.index.str.contains('Федерация') == False)]

#Выводим топ-10
display(top_in.head(10))
```

Построим график - для его вывода напишем функцию, потому что подобный график мы ещё не раз выведем в дальнейшем:

```python
def tops_plot(series, N, name):
    series_plot = series.head(N)
    plt.figure(figsize=(12, 12))
    plt.barh(series_plot.index, series_plot['indicator_value'])
    plt.title(name, fontsize=18)
    plt.xticks(fontsize=14)
    plt.ticklabel_format(style='plain', axis='x')
    plt.yticks(fontsize=14)
    plt.grid(True)
    plt.gca().invert_yaxis()
    plt.show()
```

```python
tops_plot(top_in, 10, 'Переселившиеся')
```

В топе-10 мест постоянной миграции Москва занимает почётное последнее место; а вот Московская область - первое, за девять лет туда переехали почти миллион человек. В принципе ожидаемо - неожиданными для меня вторая и третья позиции, то есть Башкортостан и Тюменская область. Санкт-Петербург в этом списке занимает почётное четвёртое место.

Сначала отделим `Общие итоги миграции по субъектам — число выбывших` и будем рассматривать всю подобную миграцию, то есть `городское и сельское население`:

У нас имеются 4 пропуска в `indicator_value` - посмотрим, где они, а также какие направления миграции у нас есть:

Для оценки эмиграции нам нужны пункты `Миграция международная`, `Миграция международная со странами СНГ` и `Миграция международная с зарубежными странами, исключая страны СНГ` - внутререгиональный пропуск Москвы нас не очень-то интересует, но можно заполнить и его. Заполнять будем средним для групп "Регион - тип миграции":

```python
moscow_set = data_out[(data_out['object_name'] == 'Москва') & 
                      (data_out['migration_direction'] == 'Миграция в пределах России внутрирегиональная')]
moscow_mean = np.floor(moscow_set['indicator_value'].mean())
data_out.loc[767055, 'indicator_value'] = moscow_mean

nen_set = data_out[(data_out['object_name'] == 'Ненецкий автономный округ') & 
                      (data_out['migration_direction'] == 'Миграция международная с зарубежными странами, исключая страны СНГ')]
nen_mean = np.floor(nen_set['indicator_value'].mean())
data_out.loc[771925, 'indicator_value'] = nen_mean

chu_set = data_out[(data_out['object_name'] == 'Чукотский автономный округ') & 
                      (data_out['migration_direction'] == 'Миграция международная с зарубежными странами, исключая страны СНГ')]
chu_mean = np.floor(chu_set['indicator_value'].mean())
data_out.loc[772439:772440, 'indicator_value'] = chu_mean

print(data_out.info())
```

Отлично, пропусков больше нет. Ради интереса можно посмотреть и топ-10 для всей миграции скопом:

```python
data_out_all = data_out[data_out['migration_direction'] == 'Вся миграция'].copy()

top_out = pd.pivot_table(data_out_all, index='object_name', values='indicator_value', aggfunc='sum').sort_values(
    by='indicator_value', ascending=False)
top_out['indicator_value'] = top_out['indicator_value'].astype('int64')

top_out = top_out[(top_out.index.str.contains('федеральный округ') == False) & 
                      (top_out.index.str.contains('Федерация') == False)]
display(top_out.head(10))
tops_plot(top_out, 10, 'Вся миграция - число выбывших')
```

Топ-3 здесь, конечно, очень интересен, учитывая, что рассматривается число выбывших - однако это вся миграция скопом, включая временную. Трактовать эти числа тяжеловато - перейдём к эмиграции:

### <center>Основные источники эмиграции</center>

Рассмотрим отдельно эмиграцию в страны СНГ; в другие страны, кроме СНГ, и эмиграцию общую - начнём с СНГ:

```python
#выделяем данные
data_out_sng = data_out[
    data_out['migration_direction'] == 'Миграция международная со странами СНГ'].copy()

#суммируем за период 2011-2019 и сортируем по убыванию
top_out_sng = pd.pivot_table(
    data_out_sng, index='object_name', values='indicator_value', aggfunc='sum').sort_values(
    by='indicator_value', ascending=False)
top_out_sng['indicator_value'] = top_out_sng['indicator_value'].astype('int64')

#убираем фед. округа и общую сумму
top_out_sng = top_out_sng[
    (top_out_sng.index.str.contains('федеральный округ') == False) & 
    (top_out_sng.index.str.contains('Федерация') == False)]
display(top_out_sng.head(10))
```

```python
tops_plot(top_out_sng, 10, 'Mиграция в страны СНГ')
```

Здесь Санкт-Петербург побеждает с очень заметным отрывом - то есть в страны СНГ в значительном количестве переехали люди из СПб. Далее неожиданно идёт Тюменская область и на третьем месте - Москва. Посмотрим эмиграцию в страны, не являющиеся странами СНГ:

```python
#выделяем данные
data_out_not_sng = data_out[
    data_out['migration_direction'] == 'Миграция международная с зарубежными странами, исключая страны СНГ'].copy()

#суммируем за период 2011-2019 и сортируем по убыванию
top_out_not_sng = pd.pivot_table(
    data_out_not_sng, index='object_name', values='indicator_value', aggfunc='sum').sort_values(
    by='indicator_value', ascending=False)
top_out_not_sng['indicator_value'] = top_out_not_sng['indicator_value'].astype('int64')

#убираем фед. округа и общую сумму
top_out_not_sng = top_out_not_sng[
    (top_out_not_sng.index.str.contains('федеральный округ') == False) & 
    (top_out_not_sng.index.str.contains('Федерация') == False)]
display(top_out_not_sng.head(10))
```

```python
tops_plot(top_out_not_sng, 10, 'Mиграция в другие страны, исключая СНГ')
```

А вот здесь побеждает Москва - СПб на почётном четвёртом месте. На втором и третьем местах - Хабаровский и Приморский край, что, конечно, крайне любопытно. Перейдём к общей эмиграции - и здесь для начала проверим, сходится ли сумма выделенных нами данных с графой `Миграция международная`:

```python
top_out_international = (top_out_not_sng + top_out_sng).sort_values(
    by='indicator_value', ascending=False) 
display(top_out_international.head(10))
```

```python
data_test = data_out[
    data_out['migration_direction'] == 'Миграция международная'].copy()

top_test = pd.pivot_table(
    data_test, index='object_name', values='indicator_value', aggfunc='sum').sort_values(
    by='indicator_value', ascending=False)

top_test = top_test[
    (top_test.index.str.contains('федеральный округ') == False) & 
    (top_test.index.str.contains('Федерация') == False)]
display(top_test.head(10))
```

Да, сходится, то есть вопросов к себе у нас нет :) Строим график:

```python
tops_plot(top_out_international, 15, 'Mиграция в другие страны')
```

Рискну предположить, что различие между Петербургом и Москвой связано в том числе с более высоким уровнем жизни в Москве, отказаться от которого людям тяжелее - в то время как жители Санкт-Петербурга в этом отношении более легки на подъём, и основные скачки питерской эмиграции - 2012 год (с `851` до `9602`) и 2014 (максимум в `62941`).

Стоит однако отметить, что Санкт-Петербург по эмиграции стал опережать Москву только в 2012 году, а в период 2000-2011 всё было наоборот, Москва СПб опережала - но в 2012 году это положение дел изменилось, причём резко. 

## <center>Выводы</center>

В Санкт-Петербург за период 2011-2019 переехало в ~1.5 раза больше людей, чем в Москву - предполагаю, что это связано с более доступными ценами на жильё. Однако если смотреть на переезды в целом, то на первом месте окажется Московская область - Москва находится на 10-м, а Санкт-Петербург - на 4-м. 2-е и 3-е занимают Башкирия и Тюменская область соответственно - почему так, сказать затрудняюсь, эта тема требует отдельного исследования.

В вопросе эмиграции в другие страны Санкт-Петербург и Москва лидируют: занимают 1-е и 2-е место соответственно, причём Санкт-Петербург побеждает с заметным отрывом (`256080` против `140947`). Эмиграция из Санкт-Петербурга носит отчётливый политический характер: её максимум (`62941`) приходится на 2014 год, а первичный резкий скачок (с `851` до `9602`) - на 2012. Москва же начиная с 2012 просто стабильно растёт - и здесь стоит отметить два момента:
+ **до** 2012 - а точнее в период **2000-2011** - Москва опережала Санкт-Петербург по миграции в другие страны; картина резко изменилась именно в **2012**: Санкт-Петербург начал опережать Москву, причём с очень заметным отрывом (исключение составляет 2017 год);
+ Санкт-Петербург является безусловным лидером по эмиграции в страны СНГ; по эмиграции в страны СНГ Москва находится только на 3-м месте - и на 1-м месте по эмиграции в страны, не являющиеся странами СНГ. Рискну предположить, что это связано с более высоким уровнем жизни в Москве, отказаться от которого людям тяжелее.

Эмиграция в Московской и Ленинградской областях по характеру совпадает с городами - однако Московская область по количеству эмигрировавших стабильно опережает Ленинградскую, начиная с 2016 года. В то время как Москва от Санкт-Петербурга отстаёт на всём отрезке 2011-2019, исключая 2017 год.
